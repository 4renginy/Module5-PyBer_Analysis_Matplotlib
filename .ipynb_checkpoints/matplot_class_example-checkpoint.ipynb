```python
#5.1.7 Scatter Plots
import matplotlib.pyplot as plt
```

```python
x= [50,42,39,34,32,48,22,17,8,13]
y=[17,19,21,22,34,23,34,31,43,35]
```

```python
plt.scatter(x,y, color="magenta", marker="x", s=x)
plt.xlabel("Total Number of rides(Per City)")
plt.ylabel("Average Fare($)")
plt.title("Average Fare vs Total Rides")
plt.show

```

```python
names = ['Tony', 'Joe', 'Sally', 'Andrea']
num_marbles = [100, 500, 150, 20]


fig, ax = plt.subplots(2,2) 
ax[0,0].plot (names,num_marbles)
ax[0,0].set_title("Line")
ax[0,1].bar (names, num_marbles)
ax[0,1].set_title("bar")
ax[1,0].scatter(names,num_marbles,s=num_marbles)
ax[1,0].set_title ("scatter")
ax[1,1].pie(num_marbles, labels=names)
ax[1,1].set_title ("pie")
plt.show()
```
